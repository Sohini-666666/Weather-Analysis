# Step 1

Importing The Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Step 2

Naming The My City Dataset And The Global Dataset With Certain Variables And Checking That Whether My Data Is Loaded Sucessfully

```python
c_dataset=pd.read_csv(r"C:\Users\HP\Desktop\udacity\my_city.csv")
g_dataset=pd.read_csv(r"C:\Users\HP\Desktop\udacity\global.csv")
```

```python
c_dataset
```

```python
g_dataset
```

# Step 3

Dropping The Missing Values

```python
c_data=c_dataset.dropna(axis = 0, how ='any')
g_data=g_dataset.dropna(axis = 0, how ='any')
```

# Step 4

Plotting My City Dataset Without Moving Average

```python
c_data.avg_temp.plot(figsize=(20,5))
plt.title("Nearest City To Me Haora Country India Without Moving Average",fontdict={'fontsize':18,'color':'Blue'})
plt.xlabel('Year',fontdict={'fontsize':16, 'color': 'Green'})
plt.ylabel('Avg_temp', fontdict={'fontsize':16,'color':'Green'})
```

Plotting My City Dataset With Moving Average

```python
c_data.avg_temp.rolling(window=7).mean().plot(figsize=(20,5))
plt.title("Nearest City To Me Haora Country India With 7-Year Moving Average",fontdict={'fontsize':18,'color':'Blue'})
plt.xlabel('Year',fontdict={'fontsize':16, 'color': 'Green'})
plt.ylabel('Avg_temp', fontdict={'fontsize':16,'color':'Green'})
```

Plotting Global Dataset Without Moving Average

```python
g_data.avg_temp.plot(figsize=(20,5))
plt.title("Global Data Without Moving Average",fontdict={'fontsize':18,'color':'Blue'})
plt.xlabel('Year',fontdict={'fontsize':16, 'color': 'Green'})
plt.ylabel('Avg_temp', fontdict={'fontsize':16,'color':'Green'})
```

Plotting Global Dataset With Moving Average

```python
g_data.avg_temp.rolling(window=7).mean().plot(figsize=(20,5))
plt.title("Global Data With 7-Year Moving Average",fontdict={'fontsize':18,'color':'Blue'})
plt.xlabel('Year',fontdict={'fontsize':16, 'color': 'Green'})
plt.ylabel('Avg_temp', fontdict={'fontsize':16,'color':'Green'})
```

```python
#storing the temperature and year columns values in new variables for plotting
year_global=g_data.get('year')
temp_global=g_data.get('avg_temp')

year_my_city=c_data.get('year')
temp_my_city=c_data.get('avg_temp')
```

# Step 5

Merging The Two Line Graphs In One Graph -> Orange Line is My City(Haora,India) And Blue Line Is Global

```python
fig = plt.figure(figsize=(20,17))
plt.subplot(5,1,1)
plt.grid(True)
plt.plot(year_global,temp_global,label='Global Temperature Over Years')
plt.plot(year_my_city,temp_my_city,label='Haora (India) Temperature Over Years')
plt.legend(loc=4)
plt.ylim(0,30)
plt.title('Haora(India) VS Global Data Temperature Over Years',fontsize=22)
plt.xlabel('Year',fontsize=17)
plt.ylabel('Average temperature in °C',fontsize=15)
```
